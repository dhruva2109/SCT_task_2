```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path_gender = 'gender_submission.csv'  # Update this with the correct file path
gender_data = pd.read_csv(file_path_gender)

# Data Cleaning: Checking for missing values and data types
print("Data Info:")
print(gender_data.info())

# Check for missing values
missing_values = gender_data.isnull().sum()
print("\nMissing Values in the Dataset:")
print(missing_values)

# Check basic statistics
print("\nBasic Statistics of the Dataset:")
print(gender_data.describe())

# Exploratory Data Analysis (EDA)

# Countplot of the 'Survived' variable
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=gender_data, palette='pastel')
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# Analysis of Survival Rate
survival_rate = gender_data['Survived'].value_counts(normalize=True) * 100
print("\nSurvival Rate (%):")
print(survival_rate)

# Further Exploration of relationships if there were more features
# If the dataset had more columns like gender, age, etc., we would explore relationships like:
# sns.countplot(x='Survived', hue='Sex', data=gender_data) for analyzing the gender effect on survival.
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path_test = '/test.csv'  # Update this with the correct file path
test_data = pd.read_csv(file_path_test)

# Data Cleaning: Checking for missing values and data types
print("Data Info:")
print(test_data.info())

# Check for missing values
missing_values = test_data.isnull().sum()
print("\nMissing Values in the Dataset:")
print(missing_values)

# Check basic statistics
print("\nBasic Statistics of the Dataset:")
print(test_data.describe())

# Handling missing values (if any)
# Example: Fill missing age with the median age
if 'Age' in test_data.columns:
    test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

# Exploratory Data Analysis (EDA)

# Plotting distribution of some numerical features
plt.figure(figsize=(10, 5))
sns.histplot(test_data['Age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Correlation heatmap (if there are numeric features)
plt.figure(figsize=(10, 6))
# Calculate correlation only for numeric columns
sns.heatmap(test_data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Further EDA
# Depending on the dataset's features, you can explore relationships
# Example: Countplot for categorical data
if 'Sex' in test_data.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Sex', data=test_data, palette='muted')
    plt.title('Count of Passengers by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path_train = '/train.csv'  # Update this with the correct file path
train_data = pd.read_csv(file_path_train)

# Data Cleaning: Checking for missing values and data types
print("Data Info:")
print(train_data.info())

# Check for missing values
missing_values = train_data.isnull().sum()
print("\nMissing Values in the Dataset:")
print(missing_values)

# Check basic statistics
print("\nBasic Statistics of the Dataset:")
print(train_data.describe())

# Handling missing values
# Fill missing 'Age' with the median age
if 'Age' in train_data.columns:
    train_data['Age'].fillna(train_data['Age'].median(), inplace=True)

# Fill missing 'Embarked' with the most common value
if 'Embarked' in train_data.columns:
    train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)

# Dropping rows where 'Cabin' is missing (optional, depending on the use case)
if 'Cabin' in train_data.columns:
    train_data.drop(columns=['Cabin'], inplace=True)

# Exploratory Data Analysis (EDA)

# 1. Distribution of Age
plt.figure(figsize=(10, 5))
sns.histplot(train_data['Age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Survived vs Not Survived (Countplot)
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=train_data, palette='pastel')
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# 3. Survived by Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=train_data, palette='muted')
plt.title('Survived by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# 4. Correlation heatmap (numeric features)
plt.figure(figsize=(10, 6))
# Calculate correlation only for numeric columns
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5) # Use numeric_only=True to consider only numeric columns for correlation calculation
plt.title('Correlation Heatmap')
plt.show()

# 5. Survived by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Pclass', data=train_data, palette='Set2')
plt.title('Survived by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# 6. Distribution of Fare
plt.figure(figsize=(10, 5))
sns.histplot(train_data['Fare'].dropna(), kde=True, bins=40)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
